```python
#Move data from RAW to Bronze
from pyspark.sql.functions import lit, current_date

# Read the CSV file from the Raw Lake House
csv_file_path = "Raw.Lakehouse/Files/car_prices.csv"
df = spark.read.csv(f"abfss://PySparkVehicleSales@onelake.dfs.fabric.microsoft.com/"+csv_file_path, header=True,  inferSchema=True)

#add Tracking Columns
df = df.withColumn('LakePath', lit(csv_file_path))
df = df.withColumn('LakeDate', current_date())

#Write the datafame to Bronze Layer as Parquet
df.write.parquet(f'abfss://PySparkVehicleSales@onelake.dfs.fabric.microsoft.com/Bronze.Lakehouse/Files/CarPrices.Parquet', 'Overwrite')

```
